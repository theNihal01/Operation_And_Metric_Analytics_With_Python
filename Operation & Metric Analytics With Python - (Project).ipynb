```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('intern-Day_01.csv')

# Group the data by branch and calculate the mean rating for each branch
branch_ratings = df.groupby('Branch')['Rating'].mean().reset_index()

# Create a bar chart of the mean ratings for each branch
plt.bar(branch_ratings['Branch'], branch_ratings['Rating'],color='lightgreen')
plt.xlabel('Branch',color='Brown',size=15)
plt.ylabel('Rating',color='orange',size=15)
plt.title('Branches with Highest and Lowest Ratings in the Dataset')

# Customize the chart to highlight the branch with the highest mean rating
highest_rating = branch_ratings['Rating'].max()
highest_branch = branch_ratings.loc[branch_ratings['Rating'] == highest_rating, 'Branch'].iloc[0]
plt.axhline(y=highest_rating, color='red', linestyle='--')
plt.text(highest_branch, highest_rating+0.1, 'Highest Rating', ha='center')

plt.show()
```

```python
# this is the very shorest form of Create Data labels :-
# And this is not any viusal of work only show to Data labels..
import matplotlib.pyplot as plt

# Data
x = ['A', 'B', 'C', 'D']
y = [10, 20, 15, 25]

# Create bar chart
plt.bar(x, y)

# Add data labels
plt.bar_label(plt.bar(x, y), labels=y)

# Show the chart
plt.show()
```

```python

months = ['Jan', 'Feb', 'Mar']
incomes = [5749.83, 4369.47, 5260.75]

plt.bar(months, incomes,color='Purple',width=0.3)

plt.bar_label(plt.bar(months, incomes), labels=incomes)

# add labels and title
plt.xlabel('Month',color='red',size=12)
plt.ylabel('Gross Income',color='green',size=18)
plt.title('Gross Income by Month',color='Brown',size=18)

plt.show()


```

```python
labels = ['Jan', 'Feb', 'Mar']
incomes = [5749.83, 4369.47, 5260.75]
colors = ['purple', 'lightblue', 'lightgreen']

plt.pie(incomes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Income Distribution by Month', color='brown', size=18)

plt.show()
```
